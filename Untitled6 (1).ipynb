```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
titanic = sns.load_dataset("titanic")
```

```python
titanic.head(1)
```

```python
# histogram /distriution of coninions numerical feature
```

```python
sns.histplot(data=titanic, x='age', bins=30, color='skyblue')
```

```python
# count plot - categorical count
```

```python
titanic['sex'].value_counts()
```

```python
sns.countplot(data= titanic ,x='sex',hue='survived', palette='husl')
```

```python
#box plot
```

```python
sns.boxplot(data=titanic, x='pclass', y='age', hue='sex', palette='viridis')
```

```python
sns.violinplot(data=titanic, x='pclass', y='age', hue='sex', palette='viridis')
```

```python
#scatter plot
```

```python
sns.scatterplot(data=titanic, x='age', y='fare', hue='sex', palette='hls')
```

```python
#heatmap - always for correlation of numerical
```

```python
sns.kdeplot(data=titanic, x='age')
```

```python
sns.pairplot(titanic)
```
