```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Boxplot - outlier finding
```

```python
plt.figure(figsize=(8,5))
```

```python
titanic = sns.load_dataset("titanic")
```

```python
titanic.head(1)
```

```python
plt.figure(figsize=(5,4))
titanic.boxplot(column= 'fare' ,by = 'pclass', grid =False)
plt.savefig("bc.jpeg")
```

```python
#scatter plot - Num vs Num -
```

```python
plt.figure(figsize=(6,4))
plt.scatter(titanic['age'],titanic['fare'], c="red")
plt.title("scatter plot of age vs fare")
plt.xlabel("age of passenger")
plt.ylabel("fare of passenger")
plt.show()
```

# charts using sns seaborn library

syntax
1 - set the figure size using plt
2 - sns.graph type name (data = df, x =column1, y=ccolumn2 , hue=column3)

3 - plt.title()

4 - plt.xlabel()

5 - plt.ylabel()

6 - plt.legend()

7 - plt.show()
